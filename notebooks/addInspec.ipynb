# Solving the godamn addresses that ain't in the api for some goddamn reason

```python
import pandas as pd
import requests
import re
```

```python
listing = pd.read_csv("../data/curated/listing_require_manual.csv")
```

```python
# Some surburbs are written as another name
# Avuenues, streets, drives can be written differently 

def addressSolv(address):
    address = re.sub(" Drive|| Street|| Avenue|| Road,|| Court|| Crescent", "", address)
    address = re.sub("^.+\/", "", address)
    address = re.sub("[0-9]+\-", "", address) 
    query = re.sub(" ", "+", address)
    query = re.sub(",", "%2C", query)
    try:
        url = f"https://nominatim.openstreetmap.org/search?q={query}&format=json&limit=1"
        location = requests.get(url).json()
        print(query)
        return location[0]["lat"], location[0]["lon"]
    except IndexError:
        #Could be that surburb is fked up, try deleting it and see 
        try:
            query = re.sub("2C\+.+\+VIC", "2C+VIC", query)
            url = f"https://nominatim.openstreetmap.org/search?q={query}&format=json&limit=1"
            location = requests.get(url).json()
            print(query)
            return location[0]["lat"], location[0]["lon"]
        except IndexError:
            #Address is bs
            return "Just do it by hand"
```

```python
listing["Coordinates"]= listing["address"].apply(lambda x: addressSolv(x))
listing["Latitude"] = listing["Coordinates"].apply(lambda x: x[0])
listing["Longitude"] = listing["Coordinates"].apply(lambda x: x[1])
```

```python
listing[listing["Coordinates"] == "Just do it by hand"]["Coordinates"]= listing[listing["Coordinates"] == "Just do it by hand"]["address"].apply(lambda x: addressSolv(x))
listing[listing["Coordinates"] == "Just do it by hand"]["Latitude"] = listing[listing["Coordinates"] == "Just do it by hand"]["Coordinates"].apply(lambda x: x[0])
listing[listing["Coordinates"] == "Just do it by hand"]["Longitude"] = listing[listing["Coordinates"] == "Just do it by hand"]["Coordinates"].apply(lambda x: x[1])
```

```python
listing.to_csv("../data/curated/listing_require_manual.csv")
```
